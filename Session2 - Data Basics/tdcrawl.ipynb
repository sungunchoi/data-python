### Column 으로 만들기

```python
# 라이브러리 선언
import pandas as pd
from bs4 import BeautifulSoup 
import requests 

#URL 선언 및 접속
targetUrl = "https://www.timeanddate.com/holidays/south-korea/"
resp = requests.get(url=targetUrl)
resp.encoding = "utf8"
html = resp.text
htmlObj = BeautifulSoup(html, "html.parser")

#태그 내 속성정보 수집
holiTable = htmlObj.find(name="table", attrs= {'id':'holidays-table'})
trTag = holiTable.findAll(name="tr")
thTag = holiTable.findAll(name="th", attrs = {'class':'nw'})
tdTag = holiTable.findAll(name="td", attrs = {'class':'nw'})
aTag = holiTable.findAll(name="a")
typeTag = holiTable.findAll(name="td")

#리스트 선언
monList = []
dayList = []
nameList = []
typeList = []

for i in range(0, len(thTag)):
    #각 태그정보 수집
    monTag = thTag[i].text
    dayTag = tdTag[i].text
    nameTag = aTag[i].text
    # 리스트에 태그 정보 append
    monList.append(monTag)
    dayList.append(dayTag)
    nameList.append(nameTag)

for i in range (2, len(typeTag),3):
    whaTag = typeTag[i].text
    typeList.append(whaTag)
    # 리스트 합치기
resultDf = pd.DataFrame(zip(monList, dayList, nameList, typeList),
                         columns=["Date"," ", "Name", "Type"]   )

resultDf
```

### Row로 만들기

```python
import requests, bs4
import pandas as pd

resp = requests.get("https://www.timeanddate.com/holidays/south-korea/")
resp.encoding='utf-8'
html = resp.text
bs = bs4.BeautifulSoup(html, 'html.parser')

tableTag = bs.find("table", {"id":"holidays-table"})

tbodyTag = tableTag.find(name = "tbody")

trs = tbodyTag.find_all(name = "tr")

columns = trs[1].find_all(name = ["th", "td"])

columns

rowList = []
columnList = []

for i in range(0, len(trs)):
    columns = trs[i].find_all(name = ["th", "td"])
    if columns == []:
        continue
    for j in range(0, len(columns)):
        columnList.append( columns[j].text)
    rowList.append(columnList)
    columnList = []
    
hoilday = pd.DataFrame(rowList, columns=["Date", "day", "Name", "Type"])
hoilday
```
