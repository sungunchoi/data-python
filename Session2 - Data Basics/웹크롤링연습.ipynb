```python
# 라이브러리 선언
import pandas as pd
from bs4 import BeautifulSoup 
import requests 

#URL 선언 및 접속
targetUrl = "http://www.ssg.com/plan/planShop.ssg?dispCmptId=6000354166"
resp = requests.get(url=targetUrl)
resp.encoding = "utf8"
html = resp.text
htmlObj = BeautifulSoup(html, "html.parser")

#리스트 선언
priceList = []
titleList = []
#태그 내 속성정보 수집
nameParent = htmlObj.findAll(name="div", attrs= {'class':'title'})
priceParent = htmlObj.findAll(name="div", attrs= {'class':'ssg_price_ko show_gl hide_ko'})

for i in range(0, len(nameParent)):
    #각 태그정보 수집
    priceTag = priceParent[i].text
    aTag = nameParent[i].find(name="a")
    titleTag = aTag.find(name="em", attrs={'class':'tx_ko'}) 
    titleTag.text
    # 리스트에 태그 정보 append
    priceList.append(priceTag)
    titleList.append(titleTag)
    
resultDf = pd.DataFrame(zip(priceList, titleList), 
                            columns=["가격","브랜드"])

resultDf
```
