### 2. 차트 꾸미기

```python
#필요라이브러리정의
import matplotlib.pyplot as plt
# 커맨드뷰에서차트시연
%matplotlib inline
x=st01Data.WEEK
y=st01Data.MEAN_VOLUME
# x축값범위지정
plt.xticks( range(min(x),max(x),3)  )
# X, Y축라벨
plt.xlabel("week")
plt.ylabel("avg_volume")
# 차트제목
plt.title("weekly trend")
# 선색상, 종류, 굵기변경
plt.plot(x,y, "bo-", lw=2.5)
```

```python
### 3. 다중차트 라벨 설정
```

```python
x=st01Data.WEEK
y=st01Data.MEAN_VOLUME
y2=st02Data.MEAN_VOLUME

plt.plot(x,y,"go--", label="ST0001")
plt.plot(x,y2,"bo-", label="ST0002")
# 각차트별legend
plt.legend(loc=0)
plt.grid(True) #그리드표기
plt.xlabel('week') # x축라벨
plt.ylabel('mean_volume') # y축라벨
plt.title('volume by week') # 차트제목
```

## 2. 시각화 고오급

```python
### 1. 서브플롯
```

```python
# 팝업창활용하여차트시연
x=st01Data.WEEK
y=st01Data.MEAN_VOLUME
y2=st02Data.MEAN_VOLUME
fig = plt.figure(figsize=(14,4))
# 첫번째차트
ax1 = fig.add_subplot(121)
ax1.plot(x,y, 'bo-',lw=1.5, label = 'ST0001')
ax1.grid(True)
ax1.legend(loc=0)
ax1.set_xlabel('ax1_index')
ax1.set_ylabel('ax1_value')
ax1.set_title('ax1_title')
# 두번째차트
ax2 = fig.add_subplot(122)
ax2.plot(x,y2, 'go-', lw=1.5, label='ST0002')
ax2.grid(True)
ax2.legend(loc=0)
ax2.set_xlabel('ax2_index')
ax2.set_ylabel('ax2_value')
ax2.set_title('ax2_title')
```

```python
### 2.이중축
```

```python
# 팝업창활용하여차트시연
x=st01Data.WEEK
y=st01Data.MEAN_VOLUME
y2=st02Data.MEAN_VOLUME
fig = plt.figure(figsize=(14,4))
# 이중축
fig = plt.figure(figsize=(14,4))
ax1 = fig.add_subplot(1,1,1)
#figure, ax1 = plt.subplots(figsize=(10,5))

ax1.plot(x,y, 'bo-',lw=1.5, label = 'ST0001')
ax1.grid(True)
ax1.legend(loc=0)
ax1.set_ylabel('ax1_value')
ax1.set_xlabel('ax1_index')
plt.title('subplot')
ax2 = ax1.twinx()
ax2.plot(x,y2, 'go-', lw=1.5, label='ST0002')
ax2.legend(loc=4)
ax2.set_ylabel('ax2_value')
```
