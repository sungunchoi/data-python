# 파트3. seaborn

### 데이터 불러오기 

```python
# Matplotlib 기반 다양한 색상 및 테마를 추가한 패키지
# 데이터 내 일반 통계함수(Im 등)구현 가능
# 다양한 변수를 효과적으로 시연가능
# 테스트를 위한 데이터 제공
```

```python
import pandas as pd
```

```python
csData = pd.read_csv("../dataset/customerdata.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
csData.columns
```

### 1. 카운트 차트 (count plot)

```python
sns.countplot(data=csData,
             x="EMI", hue="CUSTTYPE")
```

### 2. 히스토그램(dist plot)

```python
figure = plt.figure(figsize=(18,6))
ax1 = plt.subplot(1,2,1)
sns.distplot(csData.AVGPRICE, kde = False, ax=ax1 )
ax2 = plt.subplot(1,2,2)
sns.distplot(csData.AVGPRICE, hist = False, ax=ax2 )
```

### 3. 바 차트 (bar plot)

```python
sns.barplot(x="EMI", y="AVGPRICE", data = csData, hue = "CUSTTYPE", ci = "sd")
```

### 4. 박스 차트 (box plot)

```python
sns.boxplot(data=csData.loc[csData.EMI != 1], x="EMI", y="AVGPRICE")
```

### 5. lineplot

```python
# x축과 y축에 값을 할당하고 x축에 따른 y축 변수의 변화량을 시각화
```

```python
featuresData = pd.read_csv("../dataset/feature_regression_example.csv")
print(featuresData.shape)
```

```python
featuresData.head()
```

```python
sns.lineplot(data=featuresData.loc[featuresData.YEAR==2015],x="YEARWEEK", y="QTY")
```

### 6. Implot

```python
# 두 데이터의 상관관계를 linear한 형태로 뽑아내는 차트
```

```python
sns.lmplot(data=featuresData, x="QTY",y="PRO_PERCENT")
```

### 7.jointplot 구현

```python
# 두 데이터의 관계 및 각 변수의 빈도수를 같이 표현
```

```python
sns.jointplot(data=csData, x= "AVGPRICE", y = "PRODUCTAGE")
```

### 8. heatmap

```python
# 데이터의 상관관계를 확인
```

```python
#  상관분석
featuresData.corr()
```

```python
# heatmap
featuresCorr = featuresData.corr()

sns.heatmap(featuresCorr, annot = True)
```
