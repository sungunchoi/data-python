### 분산 과 표준편차

```python
st1Mean = np.mean(st1)
st1Len = len(st1)
```

```python
answer = 0
```

```python
for eachitem in st1:
    answer = answer + (st1Mean-eachitem) ** 2
```

```python
stdValue = np.sqrt(answer / (st1Len-1))
stdValue
```

# 2. matplotlib 활용 시각화 고급

### (1) 서브플롯 그리기

```python
import pandas as pd

groupDataS01 = pd.read_csv("../dataset/ST0001DATA.csv")
groupDataS02 = pd.read_csv("../dataset/ST0002DATA.csv")

# 필요 라이브러리 정의
import matplotlib.pyplot as plt
# 팝업 창 활용하여 차트 시연
x=groupDataS01.WEEK
y=groupDataS01.MEAN_VOLUME
y2=groupDataS02.MEAN_VOLUME
```

```python
fig = plt.figure(figsize=(14,4)) 
# 첫번째 차트
ax1 = fig.add_subplot(121)
ax1.plot(x,y, 'bo-',lw=1.5, label = 'ST0001') 
ax1.grid(True) 
ax1.legend(loc=0)
ax1.set_xlabel('ax1_index')
ax1.set_ylabel('ax1_value')
ax1.set_title('ax1_title')
# 두번째 차트
ax2 = fig.add_subplot(122) 
ax2.plot(x,y2, 'go-', lw=1.5, label='ST0002') 
ax2.grid(True) 
ax2.legend(loc=0)
ax2.set_xlabel('ax2_index')
ax2.set_ylabel('ax2_value')
ax2.set_title('ax2_title')
```

### (2) 이중축 그리기

```python
# import matplotlib
# font = {'family' : 'normal',
#         'weight' : 'bold',
#         'size'   : 14}
```

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 팝업 창 활용하여 차트 시연
x=groupDataS01.WEEK
y=groupDataS01.MEAN_VOLUME
y2=groupDataS02.MEAN_VOLUME

# 이중축
fig = plt.figure(figsize=(14,4))
ax1 = fig.add_subplot(1,1,1)
#figure, ax1 = plt.subplots(figsize=(10,5))
ax1.plot(x,y, 'bo-',lw=1.5, label = 'ST0001')
ax1.grid(True)
ax1.legend(loc=0)
ax1.set_ylabel('ax1_value')
ax1.set_xlabel('ax1_index')
plt.title('subplot')
ax2 = ax1.twinx()
ax2.plot(x,y2, 'go-', lw=1.5, label='ST0002')
ax2.legend(loc=4)
ax2.set_ylabel('ax2_value')
```

### [실습]
../dataset/feature_regression_example.csv
2015년도 자료만 활용하여
x축: week 이중축으로 qty, pro_percent를 
차트로 그리세요


```python
import pandas as pd

featuresData =  pd.read_csv("../dataset/feature_regression_example.csv")

featuresData2015 = featuresData[ (featuresData.YEARWEEK >= 201501) &
                                (featuresData.YEARWEEK <= 201552)]

featuresData2015.head()
```

```python
y2 = featuresData2015["PRO_PERCENT"]
print(type(y2))
```

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 팝업 창 활용하여 차트 시연
x=featuresData2015[["WEEK"]]
y = featuresData2015["QTY"]
y2 = featuresData2015[["PRO_PERCENT"]]

# 이중축
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,1,1)
plt.plot(x,y, 'bo--',lw=1.5, label = '1st')
plt.grid(True)
plt.legend(loc=2)
plt.ylabel('qty')
plt.title('subplot')
ax2 = ax1.twinx()
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.legend(loc=4)
plt.xlabel('week')
plt.ylabel('promotion')
```

# 3. Seaborn

```python
import seaborn as sns
import pandas as pd
%matplotlib inline

csData = pd.read_csv("../dataset/customerdata.csv")
csData.head(2)
```

### (1) 카운트 차트(count plot)

```python
sns.countplot(data = csData, x="EMI", hue="CUSTTYPE")
```

### (2) 히스토그램 (dist plot)

```python
figure = plt.figure(figsize=(18,6))

ax1 = plt.subplot(1,2,1)
sns.distplot(csData.AVGPRICE, kde = False, ax=ax1 )
ax2 = plt.subplot(1,2,2)
sns.distplot(csData.AVGPRICE, hist = False, ax=ax2 )
```

### (3) 바 차트 (bar plot)

```python
sns.barplot(x="EMI", y="AVGPRICE", data = csData, 
         hue = "CUSTTYPE", ci = "sd")
```

### (4) 박스 차트 (box plot)

```python
sns.boxplot(data=csData, x="EMI", y="AVGPRICE")
```

### (5) 선 차트 (line plot)

```python
import pandas as pd
featuresData = pd.read_csv("../dataset/feature_regression_example.csv")
featuresData.head()
```

```python
sns.lineplot(data=featuresData.loc[featuresData.YEAR==2015],
                                 x="WEEK", y="QTY")
```

### (6) LM 차트 (lm plot)

```python
sns.lmplot(data=featuresData, x="PRO_PERCENT",y="QTY")
```

```python
featuresData.corr()
```

### (7) 조인트 차트 (joint plot)

```python
sns.jointplot(data=csData, x= "AVGPRICE", y = "PRODUCTAGE")
```
