```python
# 1. 라이브러리 선언
import requests, bs4
import pandas as pd

# 2. 웹페이지 소스 가져오기
url = "http://www.ssg.com/search.ssg?target=all&query=%EA%B3%84%EB%9E%80"
response = requests.get(url)
response.encoding = "utf-8"
html = response.text

# 3. 필요한 태그부분 가져오기
bs = bs4.BeautifulSoup(html, 'html.parser')
ul = bs.find("ul",{"id":"idProductImg"})
lis = ul.findAll("li")

# 4. 한 행의 값을 담을 리스트와 전체 행을 담을 리스트 선언
columnList = []
dataList = []

# 5. 반복문 이용해서 상품명, 가격, 링크 정보 리스트에 담기
for li in lis:
    # 상품명
    div = li.find("div",{"class":"thmb"})
    name = div.find("img")['alt']
    columnList.append(name)
    # 상품 가격
    price = li.find("em",{"class":"ssg_price"}).text
    price = price + "원"
    columnList.append(price)
    # 상품 링크
    link = div.find("a")['href']
    columnList.append(link)
    dataList.append(columnList)
    columnList = []
dataList

# 6. 데이터프레임 만들기
productInfo = pd.DataFrame(dataList, columns = ["상품명", "상품 가격", "상품 링크"])
productInfo

# 7. 내 파일 내에 csv로 저장하기
# productInfo.to_csv("./ssgProduct.csv", encoding='ms949', index=False)    # ms949로 encoding하면 주피터 노트북에서 안 열리는 이유??
productInfo
# # 8. DB로 저장하기
# import psycopg2                       # postgreSQL 접속을 위한 라이브러리 
# from sqlalchemy import create_engine  # 데이터베이스 엔진 생성
# postgresql_engine = create_engine('postgresql://postgres:postgres@127.0.0.1:5432/postgres')

# rename = 'ssgproduct'                 # 주의! 소문자로
# productInfo.to_sql(name = rename, con = postgresql_engine, if_exists='replace', index=False)
```
