### 1. 라이브러리 선언

```python
import pandas as pd
import requests
import bs4
```

### 2. 페이지 소스 가져오기

```python
targetUrl = "https://www.timeanddate.com/holidays/south-korea/"
resp = requests.get(targetUrl)
html = resp.text
bs = bs4.BeautifulSoup(html, "html.parser")
```

### 3. 크롤링

```python
tableObj = bs.find(name="table", attrs={ "id" : "holidays-table"})
```

```python
tbodyObj = tableObj.find(name="tbody")
```

```python
rowList = []
columnList = []
```

```python
rows = tbodyObj.findAll(name = "tr")
# tr태크 내 내부태그 없는경우 continue 필요 == [] continue
```

```python
for j in range(0, len(rows)):
    columns = rows[j].findAll(name = ["th","td"])
    # 컬럼 정보 없는 경우 pass
    if columns == []:
        continue
    # 한 행의 컬럼 정보를 columnList 에 저장함
    for i in range(0, len(columns)):
        columnList.append( columns[i].text)
    # 전체 행을 담을 리스트에 columnList 수집내용 추가
    rowList.append(columnList)
    columnList = []
```

```python
answer = \
    pd.DataFrame(rowList, columns = ["date","weekinfo","holiday_name","type"] )
```
