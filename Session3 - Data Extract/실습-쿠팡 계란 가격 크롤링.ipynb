```python
# 1. 라이브러리 선언
import bs4, requests
import pandas as pd

# 2. 한 행씩 담을 리스트와 전체 행을 모두 담을 리스트 선언
columnList = []
rowList = []
```

```python
# 3. 첫페이지부터 끝페이지까지 반복문 사용해서 웹페이지 읽어오기
for i in range(0, 5):
#     3-1. 페이지가 넘어갈 때마다 page 번호만 바뀌는 것을 확인
    url = "https://www.coupang.com/np/categories/194810?page=" + str(i+1)
#     3-2. 헤더 변경 나는 크롤러가 아니라 매킨토시 mac으로 접근하고 있으니 차단하지 말아줘~ 라는 느낌임
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:71.0) Gecko/20100101 Firefox/71.0'}
#     3-3. url로 requests해서 response받기
    data = requests.get(url, headers = headers)
#     3-4. 한글 깨지지 않게 encoding해주기
    data.encoding = 'utf-8'
#     3-5. 웹페이지 원소스 html 변수로 저장
    html = data.text

# 4. BeautifulSoup 사용해서 태그로 예쁘게 정리
    bs = bs4.BeautifulSoup(html, 'html.parser')
# 5. 상품 목록부분 가져오기
    ul = bs.find("ul",{"class":"baby-product-list"})
    lis = ul.findAll("li")

# 6. 상품명과 상품가격 list에 저장하기
    for li in lis:
        name = li.find("div",{"class":"name"}).text
        name = name.replace("\n","")
    #     print(name)
        price = li.find("strong",{"class":"price-value"}).text
    #     print(price)
        columnList.append(name)
        columnList.append(price)
        rowList.append(columnList)
        columnList = []
# 7. 전체 상품정보 담은 list를 dataframe형태로 만들어주기.
productInfo = pd.DataFrame(rowList, columns = ["상품명","상품가격"])
productInfo
```
