# 7. 데이터 그룹별 집계함수

```python
import pandas as pd
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_csv("../dataset/customerdata.csv")
customerData.head()
```

무이자할부(EMI) 기준 평균지출비용(AVGPRICE)의 평균 산출

```python
# 무이자할부(EMI) 기준 평균지출비용(AVGPRICE)의 평균 산출
groupKey = ["EMI"]
meanValue = customerData.\
groupby(by=groupKey)\
        ["AVGPRICE"].\
        agg(["mean"])

# 컬럼명 변경 (mean -> MEAN_AVGPRICE)
meanValue.rename(columns={"mean":"MEAN_AVGPRICE"}, 
                 inplace=True)
# 인덱스 초기화
meanValue.reset_index(inplace=True)
meanValue
```

무이자할부(EMI) 기준 평균지출비용(AVGPRICE)의 사이즈 산출

```python
# 무이자할부(EMI) 기준 평균지출비용(AVGPRICE)의 사이즈 산출
groupKey = ["EMI"]
meanValue = customerData.\
groupby(by=groupKey)\
        ["AVGPRICE"].\
        agg(["count"])

# 컬럼명 변경 (size -> SIZE_CS)
meanValue.rename(columns={"size":"SIZE_CS"}, 
                 inplace=True)
# 인덱스 초기화
meanValue.reset_index(inplace=True)
meanValue
```

# 8. 고급 연산

```python
import pandas as pd

# 데이터 불러오기
kopoData = pd.read_csv("../dataset/kopo_product_volume.csv")

# 데이터 선택 (ST0001 제품)
kopoDataA01 = kopoData.loc[ kopoData.PRODUCTGROUP == "ST0001" ]

# 데이터 정렬
sortKey = ["REGIONID","PRODUCTGROUP","YEARWEEK"]
kopoSortedData = kopoDataA01.sort_values(sortKey, ignore_index=True)
kopoSortedData.head()
```

```python
# 이동평균 산출 (구간:5, 중심기준)
kopoSortedData["MA5"] = kopoSortedData["VOLUME"].rolling(window=5, 
                                 center=True, 
                                 min_periods=1).mean()
kopoSortedData.head()
```

```python
kopoSortedData["EFFECT"] = kopoSortedData.VOLUME / \
kopoSortedData.MA5
kopoSortedData.head()
```

```python
# 주차컬럼 생성
kopoSortedData["WEEK"] = kopoSortedData["YEARWEEK"].astype(str).str[4:6]

# 지역, 상품, 주차별 효과 산출
groupKey = ["REGIONID","PRODUCTGROUP","WEEK"]
weeklyEffect = kopoSortedData.\
                groupby(by=groupKey)["EFFECT"].\
                agg(["mean"]).reset_index()
weeklyEffect.rename(columns={"mean":"WEEKLY_EFFECT"}, inplace=True)
weeklyEffect.head(2)
```
